```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
seed = 7
dataset = pd.read_csv("pima-indians-diabetes.csv", header=None)
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,0:8]
Y = dataset.iloc[:,8]
```

```python
scalar = StandardScaler()
```

```python
X = scalar.fit_transform(X)
```

```python
# create model
model = Sequential()
model.add(Dense(150, input_dim=8, activation='relu')) # 500 neurons
model.add(Dense(1, activation='sigmoid')) # 1 output neuron
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10,  verbose=2) # 150 epoch, 10 batch size, verbose = 2
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

```

```python
# calculate predictions
predictions = model.predict(X)    # predicting Y only using X
```

```python
print(predictions)
```

```python
# Round predictions
#rounded = [int(numpy.round(x, 0)) for x in predictions]
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
